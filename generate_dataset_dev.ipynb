```python
import pandas as pd
import os
import json

os.chdir("/home/adri/Projects/phd/pipeline")
```

```python
# load frequencies
freqs_path = "data/interim/frequencies/no_sod_frequencies/merged_frequencies.tsv"
freqs = pd.read_csv(freqs_path, index_col=0)
```

```python
freqs
```

```python
# load dyn_ligtype_dict
dict_path = "data/processed/dyn_ligtype_dict.json"

with open(dict_path, "r") as f:
    dyn_ligtype_dict = json.load(f)
    
```

```python
# create a column with the ligtype based on system
freqs["ligtype"] = freqs["system"].apply(lambda x: dyn_ligtype_dict[str(x)])
```

```python
# drop unknown, biased agonist, allosteric agonist and allosteric inhibitor
freqs = freqs[freqs["ligtype"] != "unknown"]
freqs = freqs[freqs["ligtype"] != "biased agonist"]
freqs = freqs[freqs["ligtype"] != "allosteric agonist"]
freqs = freqs[freqs["ligtype"] != "allosteric inhibitor"]
```

```python
# drop replica column
freqs = freqs.drop(columns=["replica"])
```

```python
# save to csv
freqs.to_csv("data/processed/no_sod_freqs.csv", index=False)
```
